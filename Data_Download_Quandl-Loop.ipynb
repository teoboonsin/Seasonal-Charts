```python
import quandl
import pandas as pd

#Get API Key from Quandl account
API_KEY=raw_input('Please key in your API Key: ')
quandl.ApiConfig.api_key = API_KEY



```

```python


data_code_loc=raw_input('Please input the database code that identifies where the data comes from (CME, ICE): ')
data_code_ts=raw_input('Please input the product you want in database code (HO,CL,RB,B,G): ')

mth=''

###Note that RB - RBOB code is only usable for data after 2005

futures_start_month=int(raw_input('Please input the numbers 1-12 for futures delivery month - start: '))
futures_end_month=int(raw_input('Please input the numbers 1-12 for futures delivery month - end: '))
    
year=raw_input('Please input the start contract year in YYYY format: ')
year_end=raw_input('Please input the end contract year in YYYY format: ')

file_format='xls'

contracts=0

while contracts<=futures_end_month-futures_start_month:

    a=0
    
    if futures_start_month+contracts==1:
        mth='Jan'
        futures_delivery_month='F'

    elif futures_start_month+contracts==2:
        mth='Feb'
        futures_delivery_month='G'

    elif futures_start_month+contracts==3:
        mth='Mar'
        futures_delivery_month='H'

    elif futures_start_month+contracts==4:
        mth='Apr'
        futures_delivery_month='J'

    elif futures_start_month+contracts==5:
        mth='May'
        futures_delivery_month='K'

    elif futures_start_month+contracts==6:
        mth='Jun'
        futures_delivery_month='M'

    elif futures_start_month+contracts==7:
        mth='Jul'
        futures_delivery_month='N'

    elif futures_start_month+contracts==8:
        mth='Aug'
        futures_delivery_month='Q'

    elif futures_start_month+contracts==9:
        mth='Sep'
        futures_delivery_month='U'
  
    elif futures_start_month+contracts==10:
        mth='Oct'
        futures_delivery_month='V'
 
    elif futures_start_month+contracts==11:
        mth='Nov'
        futures_delivery_month='X'

    elif futures_start_month+contracts==12:
        mth='Dec'
        futures_delivery_month='Z'

    else:
        print 'You have not entered a correct number. Please re-enter: '
    

    while a<=int(year_end)-int(year):

        print a
        mydata=quandl.get(data_code_loc+'/'+data_code_ts + futures_delivery_month+str(int(year)+a))

        %cd Quandl
        %cd $data_code_ts
        %cd $mth

        mydata.to_excel(mth+" "+str(int(year)+a)+"."+file_format)

        %cd ..
        %cd ..
        %cd ..
    
        a+=1
    
    print contracts
    contracts+=1
    
  
```
